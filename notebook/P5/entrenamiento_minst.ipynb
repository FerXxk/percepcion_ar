# Introducción a las redes neuronales: identificación de caracteres conjunto MINST

En esta práctica vamos a hacer un primer entrenamiento de una red neuronal basada en perceptrón para el reconocimiento de caracteres.

En este progrema, cada carácter se considera ya segmentado. El objetivo por tanto es implementar un clasificador basado en perceptrón que permita distinguir los caracteres, reconociendo la letra que representa.

```python
# Importamos Keras

import keras
```

## Preparación del set de datos

Vamos a usar el conjunto de datos MINST, que está formado por 60000 imágenes monocromas  28x28 que contienen caracteres numéricos entre el 0 y el 9. Por ejemplo:

![Imagen MINST](minst.jpeg)

Carga y prepara el conjunto de datos [MNIST](https://keras.io/api/datasets/mnist/). 

Tenemos que convertir los datos de formato entero (entre 0 y 255) a numeros de punto flotante entre 0 y 1.

```python

mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

Construye un modelo `tf.keras.Sequential`  apilando capas. En nuestro caso, implementaremos un Perceptrón al que le añadiremos Dropout para evitar el sobreajuste. Nuestra red neuronal estará compuesta por:

- Una capa que te transforma la imagen de entrada de 28x28 a un vector de 400 componentes.
- Una capa de entrada densa con función de activación Rectified Linear Unit (relu)
. Una capa de Dropout con el 0.2 de posibilidad.
- Una capa de salida  densa con diez neuronas (una por cada carácter) con la función de activación Softmax.

Escogeremos el optimizador adam evaluancdo la precisió ndel conjunto (accuracy) y usando una función de péridas ... 

```python
model = keras.models.Sequential([
  keras.layers.Flatten(input_shape=(28, 28)),
  keras.layers.Dense(128, activation='relu'),
  keras.layers.Dropout(0.2),
  keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

 ## Entrenando y probando el modelo.

Es vital usar una parte del conjunto de datos para evaluar la red separa de la del conjunto de entrenamiento para evaluar la capacidad de generalización de la red. Así podemos detectar el sobreajuste y volver a versiones mejores de la red.

Por ello, Keras divide el dataset MINST en 60000 imágenes para entrenamiento (x_train) con sus correspondientes etiquetas (y_train).

Además, Keras separa 10000 imágenes distintas (x_test) con sus etiquetas (y_train) para evaluar el rendimiento de la red.

Procederemos a entrenar el modelo con el conjunto de entrenamiento. Posteriormente, evaluaremos el modelo en el conjunto de prueba (x_test,y_test).

```python
model.fit(x_train, y_train, epochs=5)

print("Resultados de evaluar la red en el conjunto de test:")
model.evaluate(x_test,  y_test, verbose=2)
```

El model de clasificacion de images fue entrenado y alcanzo una exactitud de ~98% en el conjunto de datos de validación.
